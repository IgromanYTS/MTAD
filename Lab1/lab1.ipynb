```python
import pandas as pd
from sklearn import preprocessing 
import scipy as sc
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sea
print("Успішно")
```

Переконуємось, що модулі успішно імпортовані

```python
data = pd.read_csv("/Users/yurab/Desktop/Python for university/MMTAD/digital_marketing_campaign_dataset.csv")
#print(digital)
print(data.shape)
data_clean = data.dropna()
data_clean.to_csv("data_cleaned.csv", index=False)
print(data_clean.shape)
#data_wide = data_clean.pivot_table(index=[ "CampaignChannel"], columns="CustomerID", values=["Age", "Income"]).reset_index()
#print(data_wide)
analize_column = ["Age", "Income", "AdSpend", "ClickThroughRate", "ConversionRate", "WebsiteVisits", "PagesPerVisit", "TimeOnSite", "SocialShares", "EmailOpens", "EmailClicks", "PreviousPurchases", "LoyaltyPoints", "Conversion"]
```

Як можна побачити, то датасет не містить жодних значень NA

Створюємо словар для збіру статистики

```python
stats = {'Column': [], 'Mean': [], 'Trimmed': [], 'Median': [], 'Var': [], 'Std': [], 'Mad': [], 'A_mad': []}
```

Шукаємо середнє значення по колонкам

```python
for column in analize_column: 
    mean = data_clean[column].mean()
    #print(mean)
    stats['Column'].append(column)
    stats['Mean'].append(mean)
    print(f"Статистика для {column}:")
    print(f"Середнє: {mean}")
    #print("\n")
#print(mean)
```

Шукаємо зрізане середнє

```python
for column in analize_column: 
    trimmed = sc.stats.trim_mean(data_clean[column], proportiontocut=0.01)
    stats['Trimmed'].append(trimmed)
    print(f"Статистика для {column}:")
    print(f"Зрізане середнє: {trimmed}")
```

Шукаємо медіану для значень

```python
for column in analize_column: 
    median = data_clean[column].median()
    stats['Median'].append(trimmed)
    print(f"Статистика для {column}:")
    print(f"Медіана: {median}")
```

Шукаємо дисперсію по стовпчикам

```python
for column in analize_column: 
    var = data_clean[column].var()
    stats['Var'].append(var)
    print(f"Статистика для {column}:")
    print(f"Дисперсія: {var}")
```

Шукаємо стандартну дисперсію по стовпчиках

```python
for column in analize_column: 
    std = data_clean[column].std()
    stats['Std'].append(std)
    print(f"Статистика для {column}:")
    print(f"Дисперсія: {std}")
```

Шукаємо медіану по стовпчиках

```python
for column in analize_column: 
    mad = data_clean[column].median()
    stats['Mad'].append(mad)
    print(f"Статистика для {column}:")
    print(f"Медіана: {mad}")
```

```python
for column in analize_column: 
    a_mad = sc.stats.median_abs_deviation(data_clean[column])
    stats['A_mad'].append(a_mad)
    print(f"Статистика для {column}:")
    print(f"Абсол. мед. відхилення: {a_mad}")
```

Імпортуємо статистичний збір

```python
statistics_df = pd.DataFrame(stats)
statistics_df.to_csv('statistics.csv', index=False)
```

Нормалізуємо набір даних

```python
d = preprocessing.normalize(data_clean[["Age", "Income", "AdSpend", "ClickThroughRate", "ConversionRate", "WebsiteVisits", "PagesPerVisit", "TimeOnSite", "SocialShares", "EmailOpens", "EmailClicks", "PreviousPurchases", "LoyaltyPoints", "Conversion"]], axis=0)
scaled_df = pd.DataFrame(d, columns=["Age", "Income", "AdSpend", "ClickThroughRate", "ConversionRate", "WebsiteVisits", "PagesPerVisit", "TimeOnSite", "SocialShares", "EmailOpens", "EmailClicks", "PreviousPurchases", "LoyaltyPoints", "Conversion"])
#scaled_df.head(10)
scaled_df
scaled_df.to_csv("scaled_df.csv", index=False)
```

```python
scaler = preprocessing.MinMaxScaler(feature_range=(0, 2))
c = scaler.fit_transform(data_clean[["Age", "Income", "AdSpend", "ClickThroughRate", "ConversionRate", "WebsiteVisits", "PagesPerVisit", "TimeOnSite", "SocialShares", "EmailOpens", "EmailClicks", "PreviousPurchases", "LoyaltyPoints", "Conversion"]])
scaled_df2 = pd.DataFrame(c, columns=["Age", "Income", "AdSpend", "ClickThroughRate", "ConversionRate", "WebsiteVisits", "PagesPerVisit", "TimeOnSite", "SocialShares", "EmailOpens", "EmailClicks", "PreviousPurchases", "LoyaltyPoints", "Conversion"])
#scaled_df2.head()
scaled_df2
scaled_df2.to_csv("scaled_df2.csv", index=False)
```

Перевіряємо стандартний розподіл (що є на графіку)

```python
shapiro_test = stats.shapiro(data_without_str)
print('Тест Шапіро-Уілка:', shapiro_test)
```

Виходить, що це є не нормальне росподілення, хоча графік перед цим казав навпаки)

Робимо вибірку для 1000 значень

```python
bootstrap=pd.DataFrame({"AdSpend":[data.sample(1000,replace=True)["AdSpend"].mean() for i in range(0,1000)]})
bootstrap
```

Обраховуємо довірчий інтервал

```python
(bootstrap["AdSpend"].quantile(0.025),bootstrap["AdSpend"].quantile(0.975))
```
